# **Estatistica Descritva**

As estatísticas descritivas envolvem os métodos usados para resumir, simplificar e descrever os principais recursos de um conjunto de dados.

Este ramo da estatística visa fornecer uma visão geral concisa e significativa dos dados por meio de medidas como tendência central (média, mediana e moda), dispersão (intervalos, variância, desvio-padrão) e representações gráficas, tais como histogramas, gráficos de barras, gráfico de setores, gráficos de dispersão e etc... .

As estatísticas descritvas nos ajudam a compreender as características fundamentais de um conjunto de dados e a torná-lo mais gerenciável para análise.

```python
# Renomeando colunas para tradução
df.rename(columns={'Unnamed: 0':'linha',
                   'Title':'titulo',
                   'Video ID':'video_id',
                   'Published At':'data_publicacao',
                   'Keyword':'palavra_chave',
                   'Likes':'qtd_curtidas',
                   'Comments':'qtd_comentarios',
                   'Views':'qtd_visualizacoes'}, inplace=True)
```

```python
# Checagem
df.info()
```

```python
# Checando a qtde de dados nulos
df.isna().sum()
```

```python
# Checando a qtde de dados nulos
df.isnull().sum()
```

```python
# Calcular a porcentagem de valores nulos em cada coluna
porcentagem_nulos = (df.isnull().sum() / len(df)) * 100

# Exibir a porcentagem de valores nulos em cada coluna
print(round(porcentagem_nulos,1))
```

```python
# Removendo linhas com valores NaN
df = df.dropna()
```

```python
# Checando a qtde de dados nulos foram eliminados
df.isnull().sum()
```

```python
df.info()
```

```python
# Verificando se os dados são únicos na coluna video_id
#Pelo fatoa de ter aparecido 'false', significa que temos dados duplicados
df.video_id.is_unique
```

```python
# Encontrando as linhas duplicadas com base na coluna video_id
df[df.duplicated(subset='video_id', keep= False)]
```

```python
# Verificando tamanho do DataFrame
df.shape
```

```python
# Eliminando duplicidades via video_id
df = df.drop_duplicates(subset='video_id')
```

```python
# Verificando
df.shape
```

```python
# Verificando
print('A coluna video_id possui valores unicos? Resposta: ', df.video_id.is_unique)
```

```python
# Verificando valores unicos na coluna 'titulo
# Como a resposta foi false, significa que temos titulos iguais na coluna
print('A coluna titulo possui valores unicos? Resposta: ', df.titulo.is_unique)
```

```python
# Encontrar linhas duplicadas com base na coluna 'titulo'
df[df.duplicated(subset='titulo', keep=False)]
```

```python
# Eliminação de colunas desnecessárias primeiramente
# O parâmetro axis=1 indica que as colunas devem ser removidas
# (em oposição a linhas, que seria axis=0).
df.drop(['linha',
         'video_id'], axis=1, inplace=True)
```

```python
# Convertendo valores numéricos de float para int
df['qtd_curtidas'] = df['qtd_curtidas'].astype(int)
df['qtd_comentarios'] = df['qtd_comentarios'].astype(int)
df['qtd_visualizacoes'] = df['qtd_visualizacoes'].astype(int)
```

```python
# Convertendo a coluna de data_publicação para datetime
df['data_publicacao'] = pd.to_datetime(df['data_publicacao'])
```

```python
# Checagem da conversão
df.dtypes
```

```python
# Separando dia, mes e ano em colunas diferentes
df['ano'] = df.data_publicacao.dt.year

df['mes'] = df.data_publicacao.dt.month

df['dia'] = df.data_publicacao.dt.day
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Converter as colunas para int
df['ano'] = df['ano'].astype(str)
df['mês'] = df['mes'].astype(str)
df['dia'] = df['dia'].astype(str)
```
